```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad,unpad
from Crypto.Random import get_random_bytes

key=get_random_bytes(16) #128 bitlik anahtar

def sifrele(data,key):
    cipher=AES.new(key,AES.MODE_CBC) #CBC modunda AES şifreleme oluşturma uzun metinleri de şifreler
    iv=cipher.iv #başlangıç vektörü

    encrypted_data=cipher.encrypt(pad(data,AES.block_size))
    return iv + encrypted_data

def sifre_coz(encrypted_data,key):
    iv=encrypted_data[:16]
    cipher=AES.new(key,AES.MODE_CBC,iv)
    decrypted_data=unpad(cipher.decrypt(encrypted_data[16:]),AES.block_size)
    return decrypted_data

data="Mesajımın şifrelenemesi gerekiyor.Mesajım şu:Ayşe tatile çıksın!".encode("utf-8")
print("Orjinal Veri:",data)


#sifrele
encrypted_data = sifrele(data,key)
print("Şifrelenmiş veri:",encrypted_data)

#şifreyi çöz
decrypted_data=sifre_coz(encrypted_data,key)
print("şifresi çözülmüş veri:",decrypted_data.decode("utf-8"))
```
