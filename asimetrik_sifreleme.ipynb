```python
from Crypto.Cipher import PKCS1_OAEP
from Crypto.PublicKey import RSA
from Crypto.Random import get_random_bytes
from Crypto.Hash import SHA256
from base64 import b64encode,b64decode

#anahtar oluşturma
def generate_rsa_keypair():
  key=RSA.generate(2048)
  private_key = key.export_key() #private key
  public_key=key.publickey().export_key() #public key oluşturuldu
  return private_key,public_key

def encrypt_message(public_key,message):
  rsa_key=RSA.import_key(public_key)
  cipher=PKCS1_OAEP.new(rsa_key,hashAlgo=SHA256)
  encrypted_message=cipher.encrypt(message)
  return b64encode(encrypted_message)

def decrypt_message(private_key,encrypted_message):
  rsa_key=RSA.import_key(private_key)
  cipher=PKCS1_OAEP.new(rsa_key,hashAlgo=SHA256)
  decrypted_message=cipher.decrypt(b64decode( encrypted_message))
  return decrypted_message




print("public key:",public_key.decode())

message="Bu bir gizli mesajdır!".encode("utf-8")
encrypted_message=encrypt_message(public_key,message)

print("şifreli mesaj:",encrypted_message.decode())

decrypted_message=decrypt_message(private_key,encrypted_message)
print("çözülmüş mesaj:",decrypted_message.decode("utf-8")
      
      )
```
